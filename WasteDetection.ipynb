```python
import matplotlib.pyplot as plt
import numpy as np 
import cv2
import os
```

```python
DATADIR = "E:/Waste-Detection/data/train"
CATAGORIES = ["glass", "plastic", "cardboard", "paper"]
    
for category in CATAGORIES:
    path= os.path.join(DATADIR, category) # paths to the plastic, glass cardboard, metal and paper directory
    for images in os.listdir(path):
        image_array = cv2.imread(os.path.join(path, images),cv2.IMREAD_GRAYSCALE)
        plt.imshow(image_array,cmap="gray")
        plt.show()
        break
    break
    
```
